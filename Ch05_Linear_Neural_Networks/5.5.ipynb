```python
%matplotlib inline
import sys
sys.path.insert(0, '..')
import d2l
from IPython import display
from matplotlib import pyplot as plt
import torch
import torchvision
from torchvision import transforms
from torch.utils.data import DataLoader

import time
```

PyTorch의 `torchvision.datasets`는 많은 벤치마트 vision dataset에 쉬운 접근을 제공한다.  
1. `data.vision.FashionMNIST(train=True)`를 호출해서 training data를 수집한다.
2. 결과적으로 PyTorch는 인터넷 연결을 통해 다운로드된 local copy를 사용한다.
`train`파라미터의 값을 `True`나 `False`로 설정해서 training set이 필요한지, test set이 필요한지를 명시한다.

```python
trans = transforms.ToTensor() 
```

`ToTensor()`를 통해 PIL type을 Tensor로 바꿔준다

```python
mnist_train = torchvision.datasets.FashionMNIST(root="./", train=True, transform=trans, target_transform=None, download=True)
mnist_test = torchvision.datasets.FashionMNIST(root="./", train=False, transform=trans, target_transform=None, download=True)
```

10개의 각 카테고리의 training set에는 6,000개의 이미지, test set에는 1,000개의 이미지가 있다.

```python
len(mnist_train), len(mnist_test)
```

First dimension은 channel의 개수이다. dataset이 grayscale 이미지이기 때문에, channel의 갯수는 1이다. color image의 경우에는 R, G, B의 3 채널이 있다.

각 이미지의 label은 64비트 정수(스칼라)로 나타내어진다.

```python
def get_fashion_mnist_labels(labels):
    text_labels = ['t-shirt', 'trouser', 'pullover', 'dress', 'coat',
                   'sandal', 'shirt', 'sneaker', 'bag', 'ankle boot']
    return [text_labels[int(i)] for i in labels]
```

Fashion-MNIST의 10가지 카테고리

```python
def show_fashion_mnist(images, labels):
    display.set_matplotlib_formats('svg')
    # _는 변수를 사용하지 않는다는 것을 의미한다.
    _, figs = plt.subplots(1, len(images), figsize=(12, 12))
    for f, img, lbl in zip(figs, images, labels):
        f.imshow(img.reshape((28, 28)).numpy())
        f.set_title(lbl)
        f.axes.get_xaxis().set_visible(False)
        f.axes.get_yaxis().set_visible(False)
```

```python
X=[]
y=[]
for idx, data in enumerate(mnist_train):
    if(idx >= 0 and idx < 10):
        X.append(data[0])
        y.append(data[1])
    if (idx >= 10):
        break
# X, y = mnist_train[0:9]
show_fashion_mnist(X, get_fashion_mnist_labels(y))
```

※ PyTorch DataLoader object들은 배열 슬라이싱을 지원하지 않기 때문에, iterate 하여야한다.

```python
batch_size = 256
if sys.platform.startswith('win'):
    # set 0 for windows
    # 0 means no additional processes are needed to speed up the reading of data
    num_workers = 0
else:
    num_workers = 4

train_iter = DataLoader(mnist_train, batch_size, shuffle=True, num_workers=num_workers)
test_iter = DataLoader(mnist_test, batch_size, shuffle=False, num_workers=num_workers)
```

dataset을 획득하고 읽어들이는 로직은 `d2l.load_data_fashion_mnist` 함수에 캡슐화되어있고, 이후의 챕터에서 사용해 볼 것이다.
